# Summary Statistics and Visualizations

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style('whitegrid')
```

```python
df = pd.read_csv('../data/weight-height.csv')
df.head()
```

```python
# shape
df.shape
```

## Mean, Median, Mode, Min, Max and Quantiles

```python
# summary statistics / numerical summary 
df.describe()
```

## Univariate Analysis

```python
df['Gender'].value_counts() 
```

```python
df['Gender'].value_counts(normalize=True) * 100 
```

```python
sns.countplot(df['Gender'])
plt.show() 
```

### Considering only Height

```python
height = df['Height']
height.head()
```

```python
height.shape
```

## Histogram

```python
sns.distplot(height)
plt.show() 
```

```python
sns.distplot(height, kde=False) 
plt.show() 
```

```python
sns.distplot(height, rug=True, kde=False)
plt.show() 
```

```python
plt.figure(figsize=(10,6))
height.plot(kind = 'hist',title = 'Histogram of Height')
plt.show() 
```

```python
plt.figure(figsize=(10,6))
height.plot(kind = 'hist',title = 'Histogram of Height', bins=30)
plt.show() 
```

## Box-Plot

```python
plt.figure(figsize=(10,6))
height.plot(kind = 'box',
            title = 'Height Box-plot')

plt.show() 
```

## KDE Distribution

```python
height.plot(kind = 'kde',
            title = 'Height KDE', figsize=(10,6))
plt.show() 
```

## Bivariate Analysis

#### Considering both height and weight

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = "Height", y = "Weight", data=df)
plt.show() 
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = "Height", y = "Weight", data=df, hue='Gender')
plt.show()
```

```python
gender_groupby = df.groupby('Gender', as_index=False)
gender_groupby.head()
```

```python
gender_groupby.describe().T
```

## Distribution Plots

#### Considering both gender and height

```python
sns.FacetGrid(df, hue = 'Gender', height = 5).map(sns.distplot, 'Height').add_legend()
plt.show() 
```

#### Considering both gender and weight

```python
sns.FacetGrid(df, hue = 'Gender', height = 5).map(sns.distplot, 'Weight').add_legend()
plt.show() 
```

## Violin Plot

#### Gender vs Height

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = 'Gender', y ='Height', data = df)
plt.show() 
```

#### Gender vs Weight

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = 'Gender', y ='Weight', data = df)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x = 'Gender', y ='Height', data = df)
plt.show() 
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x = 'Gender', y ='Weight', data = df)
plt.show() 
```
