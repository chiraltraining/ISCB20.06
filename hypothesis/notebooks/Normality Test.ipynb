## Shapiro-Wilk Test

```python
# Set Test 
stat, p = stats.shapiro(df['Height'])

print(f'Statistics={stat}, p={p}')

# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian/Normal Distribution (fail to reject H0)')
else:
	print('Sample does not look Gaussian/Normal Distribution (reject H0)')
```

```python
# Set Test 
stat, p = stats.shapiro(df['Weight'])

print(f'Statistics={stat}, p={p}')

# interpret
alpha = 0.05
if p > alpha:
	print('Sample looks Gaussian/Normal Distribution (fail to reject H0)')
else:
	print('Sample does not look Gaussian/Normal Distribution (reject H0)')
```
