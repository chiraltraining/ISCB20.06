```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
data = np.linspace(1, 10, 50)
```

```python
data
```

```python
plt.style.available
```

## Line Plot 

```python
# Figure size 
plt.figure(figsize=(10,6))
plt.style.use( 'fivethirtyeight')
# create plot 
plt.plot(np.sin(data), label="Sin", marker='o', linewidth=2, color='b')
plt.plot(np.cos(data), label="Cos", marker='s', linewidth=2.5, color='r')

# axis-label 
plt.xlabel("X-Label", fontsize=14)
plt.ylabel("Y-Label", fontsize=14)

# title 
plt.title("Sin and Cosine Curve", fontsize=14)

# show legend 
plt.legend(loc='best') 

# export 
# plt.savefig("sin_cosin.jpeg", dpi=200)

plt.tight_layout() 

# show
plt.show() 
```

```python
fig, (ax1, ax2)  = plt.subplots(nrows=1, ncols=2)

ax1.plot(np.sin(data))
ax1.set_xlabel("X-Label")
ax1.set_ylabel("Y-Label")

ax2.plot(np.cos(data))
ax2.set_xlabel("X-Label")
ax2.set_ylabel("Y-Label")


plt.show()
```
